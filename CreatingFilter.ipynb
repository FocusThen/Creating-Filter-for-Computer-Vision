```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('../images/city_hall.jpg')

image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
```

```python
gray_image = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_image, cmap="gray")
```

```python
#Sobel filter
sobel_x = np.array([[-1,0,1],
          [-2,0,2],
          [-1,0,1]])
filter_image = cv2.filter2D(gray_image, -1, sobel_x)
```

```python
plt.imshow(filter_image, cmap="gray")
plt.title('Filtered image')
```

```python
#create binary image

retval, binary_image = cv2.threshold(filter_image, 100, 255, cv2.THRESH_BINARY)

plt.imshow(binary_image, cmap="gray")
```
